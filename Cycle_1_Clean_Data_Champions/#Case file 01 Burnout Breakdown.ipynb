```python
import pandas as pd

df = pd.read_csv('mental_health_workplace_survey.csv')

df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.isnull().sum()
```

```python
df[df['WorkHoursPerWeek'] > 18]
```

```python
df=df[df['WorkHoursPerWeek']<=18]
```

```python
median_hours = df['WorkHoursPerWeek'].median()
df.loc[df['WorkHoursPerWeek']>18, 'WorkHoursPerWeek']=median_hours
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.scatterplot(x='SleepHours', y='StressLevel', data=df)
plt.title("Stress vs Sleep")
plt.xlabel("Sleep Hours")
plt.ylabel("Stress Level")
plt.show()
```

```python
sns.boxplot(x='MentalHealthDaysOff', y='BurnoutLevel', data=df)
plt.title("Burnout Level vs Mental Health Days Off")
plt.xlabel("Mental Health Days Off")
plt.ylabel("Burnout Level")
plt.show()
```

```python
sns.scatterplot(x='HasMentalHealthSupport', y='BurnoutLevel', data=df)
plt.title("Burnout vs Mental Health support")
plt.xlabel("Mental Health Support")
plt.ylabel("Burnout Level")
plt.show()
```

```python
df.to_csv("cleaned_mental_health_workplace_survey.csv", index=False)
```

```python
#insights
1.Employees who sleep less than 5 hours will have more stress levels.
2.Better mental health support appears to reduce burnout risk.
```
